# 23 - RAG Evaluation Playground

```python
from agentic_assistants.vectordb.base import VectorStore

store = VectorStore.create()
queries = ['How do templates work?', 'How do traces sync?']
for q in queries:
    hits = store.search(q, collection='docs', top_k=5)
    scores = [h.score for h in hits[:3]]
    print({'query': q, 'hits': len(hits), 'top_scores': scores})
```
