# 17 - LLM Document Augmentation

```python
import httpx
from agentic_assistants.data.rag.augmenters import SummaryAugmenter
from agentic_assistants.data.rag.chunkers import TextChunker

def llm(prompt: str) -> str:
    r = httpx.post('http://localhost:11434/api/generate', json={'model':'llama3.2','prompt':prompt,'stream':False}, timeout=60.0)
    r.raise_for_status()
    return r.json().get('response', '')

chunk = TextChunker(chunk_size=200, chunk_overlap=0).chunk('LLM augmentation can compress context for retrieval and generation.')[0]
print(SummaryAugmenter(use_llm=True, llm_fn=llm).augment(chunk, {}).metadata)
```
