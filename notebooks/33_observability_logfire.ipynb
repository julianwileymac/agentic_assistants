## About this topic

Logfire builds on OpenTelemetry: spans nest work units, attributes carry dimensions, and exporters can target hosted collectors or files. Framework instrumentors auto-create spans around handlers and queries; manual spans remain useful for LLM calls, retrieval steps, and custom business events.
