# 22 - LanceDB vs Chroma Benchmark

```python
import time
from agentic_assistants.vectordb.base import VectorStore

for backend in ['lancedb', 'chroma']:
    store = VectorStore.create(backend=backend)
    t0 = time.time()
    _ = store.search('benchmark query', collection='benchmark', top_k=5)
    print(backend, round((time.time() - t0) * 1000, 2), 'ms')
```
