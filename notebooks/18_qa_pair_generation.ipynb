```python
from agentic_assistants.data.rag.augmenters import QAGenerationAugmenter
from agentic_assistants.data.rag.chunkers import TextChunker

def fake_llm(prompt: str) -> str:
    return 'What is the key idea?\nHow does this improve retrieval?\nWhen should it be used?'

chunk = TextChunker(chunk_size=200, chunk_overlap=0).chunk('Question generation gives retrieval more entry points.')[0]
print(QAGenerationAugmenter(num_questions=3, llm_fn=fake_llm).augment(chunk, {}).metadata)
```
