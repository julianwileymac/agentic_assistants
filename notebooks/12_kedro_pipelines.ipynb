# Kedro-Inspired Data Pipelines

This notebook demonstrates the Kedro-inspired data engineering features added to the Agentic Assistants framework.

## Features Covered

1. **DataCatalog** - YAML-based dataset configuration with versioning
2. **Pipelines & Nodes** - Pure function-based data processing
3. **Runners** - Sequential, parallel, and threaded execution
4. **Data Layers** - Convention-based data organization
5. **Hooks** - Lifecycle event handlers (MLFlow, validation, telemetry)
6. **Configuration** - OmegaConf-style config loading

```python
# Import the core components
from agentic_assistants.data.catalog import DataCatalog
from agentic_assistants.pipelines import Pipeline, node
from agentic_assistants.pipelines.runners import SequentialRunner
from agentic_assistants.data.layers import DataLayerManager, DataLayer
from agentic_assistants.hooks import HookManager
from agentic_assistants.engine import AgenticEngine

import pandas as pd
import numpy as np
```

## 1. Data Catalog

The DataCatalog provides Kedro-style dataset management with YAML configuration.

```python
# Create a DataCatalog from YAML configuration
# catalog = DataCatalog.from_yaml("../conf/base/catalog.yaml")

# Or create programmatically
catalog = DataCatalog()

# Add datasets
catalog.add_dataset('raw_users', {
    'type': 'csv',
    'filepath': '../data/01_raw/users.csv',
})

print(f"Datasets in catalog: {catalog.list_datasets()}")
```

## 2. Pipelines & Nodes

Define data processing as pure functions wrapped in nodes.

```python
# Define node functions
def clean_data(df: pd.DataFrame) -> pd.DataFrame:
    """Clean and validate data."""
    return df.dropna().drop_duplicates()

def add_features(df: pd.DataFrame) -> pd.DataFrame:
    """Add derived features."""
    df = df.copy()
    # Add your feature engineering here
    return df

# Create pipeline
data_pipeline = Pipeline([
    node(clean_data, inputs="raw_data", outputs="cleaned_data", tags={"preprocessing"}),
    node(add_features, inputs="cleaned_data", outputs="featured_data", tags={"feature_engineering"}),
])

print(data_pipeline.describe())
```

## 3. Using the AgenticEngine

The AgenticEngine provides a unified interface for all features.

```python
# Create the AgenticEngine
engine = AgenticEngine()

# Access the new Kedro-inspired features
print(f"Catalog datasets: {engine.list_datasets()}")
print(f"Pipelines: {engine.list_pipelines()}")

# Register and run a pipeline
engine.pipeline_registry.register("my_pipeline", data_pipeline)
print(f"After registration: {engine.list_pipelines()}")

# Run pipeline (uncomment when you have data)
# result = engine.run_pipeline("my_pipeline", runner="sequential")
# print(f"Success: {result.success}")
```
